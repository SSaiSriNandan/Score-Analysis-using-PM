```python
for i in range(len(df.index)):
    ele = df.loc[i]
    for j in range(1,4):
        el = []
        el.append(ele[0])
        ##el.append(pd.to_datetime(date[j]))
        el.append(date[j])
        el.append('attandance')
        sstr1='Month'+str(j)
        el.append(sstr1)
        #el.append(ele[j])
        if(ele[j]>=90):
            str1="G90"
            v=90
        elif(ele[j]>=75):
            str1="G75"
            v=75
        elif(ele[j]>=50):
            str1="G50"
            v=50
        elif(ele[j]<50):
            str1="L50"
            v=40
        el.append(v)
        el.append(sstr1+str1)
        el.append(str1)
        fres.append(el)
```

```python
for i in range(len(df.index)):
    mele = mdf.loc[i]
    for j in range(1,4):
        mel = []
        mel.append(mele[0])
        ##mel.append(pd.to_datetime(mdate[j]))
        mel.append(mdate[j])
        mel.append('marks')
        sstr2='CIE-'+str(j)
        mel.append(sstr2)
        #mel.append(mele[j]*10)
        if(mele[j]*10>=90):
            str2="G90"
            v=90
        elif(mele[j]*10>=75):
            str2="G75"
            v=75
        elif(mele[j]*10>=50):
            str2="G50"
            v=50
        elif(mele[j]*10<50):
            str2="L50"
            v=40
        mel.append(v)
        mel.append(sstr2+str2)
        mel.append(str2)
        fres.append(mel)
    for j in range(5,10):
        mel = []
        mel.append(mele[0])
        mel.append(mdate[j])
        mel.append('marks')
        if(j==5):
            sstr3=('asst')
            val=(mele[j]*10)
        elif(j==6):
            sstr3=('quiz')
            val=(mele[j]*10)
        elif(j==7):
            sstr3=('internals')
            val=(mele[j]*10/3)
        elif(j==8):
            sstr3=('see')
            val=(mele[j]*2)
        elif(j==9):
            sstr3=('finals')
            val=(mele[j])
        mel.append(sstr3)
        #mel.append(val)
        if(val>=90):
            str3="G90"
            v=90
        elif(val>=75):
            str3="G75"
            v=75
        elif(val>=50):
            str3="G50"
            v=50
        elif(val<50):
            str3="L50"
            v=40
        mel.append(v)
        mel.append(sstr3+str3)
        mel.append(str3)
        fres.append(mel)
```
